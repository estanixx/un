```python
import pandas as pd

def get_sells(i) -> pd.DataFrame:
    df = pd.read_excel(f'./resources/{i}v.xlsx', skiprows=13, usecols='B:J')
    df['Fecha'] = pd.to_datetime(df['Fecha'], 'ignore')
    df = df.loc[df['Tipo'] == 'Venta']
    df = df.dropna()
    df = df.drop(columns=['Vendedor', 'Contacto', 'Categoría de gasto', 'Estado', 'Tipo'])
    df['Sede'] = i 
    return df

df = pd.concat([get_sells(i) for i in range(1, 4)], ignore_index=True)
df
```

```python
df.to_csv('public/ventas.csv')
```
