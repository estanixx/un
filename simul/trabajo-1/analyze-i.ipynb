```python
import pandas as pd
def get_inventario(i) -> pd.DataFrame:
    df = pd.read_excel(f'resources/{i}i.xlsx', usecols='B:J', skiprows=7)
    df = df.drop(columns=['Categoría', 'Notas', 'Costo unitario', 'Costo inventario', 'Precio inventario'])
    df['Creado'] = pd.to_datetime(df['Creado'], 'ignore')
    df['Sede'] = i
    return df

df = pd.concat([get_inventario(i) for i in range(1, 4)], ignore_index=True)
df
```

```python
df.to_csv('public/inventario.csv')
```
