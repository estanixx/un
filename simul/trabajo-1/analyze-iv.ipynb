```python
import pandas as pd
def process_1iv_mayo():
    df = pd.read_excel('./resources/1iv.xlsx', skiprows=3, usecols='B:C', nrows=33, sheet_name='INFORME MAYO 2025 ')
    df = df.dropna()
    df = df.rename(columns={'GASTO': 'Descripción', 'VALOR': 'Valor'})
    df['Mes'] = 5
    df['Sede'] = 1
    return df

def process_1iv_abril():
    df = pd.read_excel('./resources/1iv.xlsx', skiprows=3, usecols='B:C', nrows=39, sheet_name='INFORME ABRIL 2025')
    df = df.dropna()
    df = df.rename(columns={'GASTO': 'Descripción', 'VALOR': 'Valor'})
    df['Mes'] = 4
    df['Sede'] = 1
    return df

def process_1iv_marzo():
    df = pd.read_excel('./resources/1iv.xlsx', skiprows=3, usecols='B:C', nrows=27, sheet_name='INFORME MARZO 2025')
    df = df.dropna()
    df = df.rename(columns={'GASTO': 'Descripción', 'VALOR': 'Valor'})
    df['Mes'] = 3
    df['Sede'] = 1
    return df

def process_1iv_feb():
    df = pd.read_excel('./resources/1iv.xlsx', skiprows=3, usecols='B:C', nrows=26, sheet_name='INFORME FEBRERO 2025')
    df = df.dropna()
    df = df.rename(columns={'GASTO': 'Descripción', 'VALOR': 'Valor'})
    df['Mes'] = 2
    df['Sede'] = 1
    return df

def process_1iv_ene():
    df = pd.read_excel('./resources/1iv.xlsx', skiprows=3, usecols='B:C', nrows=27, sheet_name='INFORME ENERO 2025')
    df = df.dropna()
    df = df.rename(columns={'GASTO': 'Descripción', 'VALOR': 'Valor'})
    df['Mes'] = 1
    df['Sede'] = 1
    return df
```
